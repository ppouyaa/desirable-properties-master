## Getting the results

```python
import numpy as np
import pandas as pd
from functools import reduce

results =[]
for i in range(10):
    results.append([pd.read_csv("new_results2_it_"+str(i)+".csv")])

_sum = reduce(lambda a, b: a.add(b, fill_value=0), [results[0][0], results[1][0], results[2][0], results[3][0],\
    results[4][0], results[5][0], results[6][0], results[7][0], results[8][0], results[9][0]])
```

```python
_sum = _sum.drop(columns = [ "num_obj", "iteration", "reference_point"])
average = _sum.drop(columns="problem") / 10
```

```python
average.to_csv("average.csv")
```
